```python
# from df1 we collect portscan target samples and from df0 we collect
df1=df1.loc[df1[' Label']=='PortScan']
df0=df0.loc[df0[' Label']=='DDoS']
df2 = df7.loc[df7[' Label']=='DoS Slowhttptest']
```

```python
# concatenating df5,df1,df0 files that has what targets we need
df=pd.concat(objs=[df5,df1,df0,df2],ignore_index=True)
df.head()
```

```python
# "df" is our concatenated data
df.shape
```

```python
# checking the available targets in the df file
df[' Label'].unique()
```

```python
# checking the available samples in each target
df[' Label'].value_counts()
```

```python
# we removing the 'Web Attack  Sql Injection' because of less samples size
df=df.loc[df[' Label']!='Web Attack  Sql Injection']
```

```python
# checking the available samples in each target
df[' Label'].value_counts()
```

```python
class_dict={}
for idx,label in enumerate(df[' Label'].unique().tolist()):
    class_dict[label]=label
print(class_dict)
```

```python
# renaming the target names to handle easily
class_dict={'BENIGN': 'BENIGN', 'Web Attack  Brute Force': 'Brute Force', 'Web Attack  XSS': 'XSS', 'PortScan': 'PortScan', 'DDoS': 'DDoS', 'DoS Slowhttptest':'DoS'}
```

```python
df['Label']=df[' Label'].map(class_dict)
```

```python
df=df.drop(labels=' Label',axis=1)
```

```python
df['Label'].value_counts()
```

```python
# we removing the 'Web Attack  Sql Injection' because of less samples size
df=df.loc[df['Label']!='XSS']
```

```python
# present targets and size of targets
df['Label'].value_counts()
```
